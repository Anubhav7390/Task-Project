```python
ax = sns.barplot(x = "Segment", y = "Sales", data = df, errorbar = None)
ax.bar_label(ax.containers[0], fmt = "%.2f", label_type = "center", color = 'black')
plt.title("Sales by Segment")
plt.show()
```

```python
ax = sns.barplot(x = "Region", y = "Sales", data = df, errorbar = None)
ax.bar_label(ax.containers[0], fmt = "%.2f", label_type = "center", color = 'black')
plt.title("Sales by Region")
plt.show()
```

```python
ax = sns.barplot(x = "Ship Mode", y = "Sales", data = df, errorbar = None)
ax.bar_label(ax.containers[0], fmt = "%.2f", label_type = "center", color = 'black')
plt.title("Sales by Ship Mode")
plt.xticks(rotation = 45)
plt.show()
```

```python
sns.barplot(x = "Sub-Category", y = "Sales", data = df, errorbar = None)
plt.title("Sales by Sub-Category")
plt.xticks(rotation = 75)
plt.show()
```
